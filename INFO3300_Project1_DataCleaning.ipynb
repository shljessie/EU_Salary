```python
fudev=['FullStack_Developer','Full Stack','fullstack','Full stack','Fullstack','Full-stack','fullsteck']
bdev= ['Devops Engineer','Senior software engineer ','System Engineer','Senior software engineer','Computer Vision','Mobile developer','SharePoint Developer','Android dev','C++ sr. s/w dev','Java','Android','Oracle Pl/SQL Developer','Senior .Net developer','Go Developer','Senior PHP developer','Hardware','SharePoint','ios','C++','Scala','IOS','Backend',' .Net Developer','Senior/Lead Developer','Scala Developer','Mobile App Developer','PHP Developer','Android Developer','C++ developer','BackEnd_Developer','Software','Java','QA','SW','Automation QA','Embedded engineer','Senior Software Engineer','DevOps','Senior DevOps','Test Automation Engineer','Senior PHP Magento developer ','E.g. C# Developer','Low level software developer','QA Ingenieur','Ruby','ruby','Java Developer ','C# Developer ','DevOps Architect ','Quality engineer','Senior Automation Test Engineer','Product Security Engineer','Lead Java Developer','Solution Engineer','Software Engineer','Erlang Developer','Oracle Developer','Lead developer','Softwaredeveloper','Senior Automation Test Engineer','System Architect','.NET','iOS','C++ Software','Python','python','Pyhon','Senior Automation Test Engineer','backend','Golang','Computer Vision Intern','Php','C#','Senior Java Developer','Java Developer','IT Engineer','.net developer','.net c# ','#c','Backend Developer']
fdev=['Senior JS Enigneer','Front-end','Sr. JS Developer','web','FrontEnd_Developer','Front End Developer ','Front End developer','Front End Engineer','Front-end developer ','FrontEnd Developer','Frontent Developer','FrontEnd dev','Front-end developer','webdev','Senior Android Developer','Frontend','Web Developer','frontend','Javascript Developer']
datas=['Senior SQL Server DBA/Developer','Business Analyst','Data Scientist','Data','data','Data science','data scientist','Data Science','Data analyst','Data Developer','Data Processing']
ux=['UX_Designer','UX_Designer','Designer','UX','ux','Ux','UI Designer','UX/UI designer','Graphic Designer','Designer','UI Designer','PCB Designer']
```

```python
data2['Position'].unique()
```

```python
positions=['UX_Designer','FullStack_Developer','BackEnd_Developer','FrontEnd_Developer','Data Scientist']
```

```python
for i in range(len(data2)):
  print(i)
  for x in range(len(ux)):
    if ux[x] in data2['Position'][i]:
      data2['Position'][i]="UX_Designer"
      print(data2['Position'][i])

  for y in range(len(fudev)):
    if fudev[y] in data2['Position'][i]:
      data2['Position'][i]="FullStack_Developer"
      print(data2['Position'][i])

  for t in range(len(bdev)):
    if bdev[t] in data2['Position'][i]:
      data2['Position'][i]="BackEnd_Developer"
      print(data2['Position'][i])
  
  for p in range(len(fdev)):
    if fdev[p] in data2['Position'][i]:
      data2['Position'][i]="FrontEnd_Developer"
      print(data2['Position'][i])
  
  for w in range(len(datas)):
    if datas[w] in data2['Position'][i]:
      data2['Position'][i]="Data Scientist"
      print(data2['Position'][i])


```

```python
data2 = data2[data2.Position.isin(positions)]
data2.head()
```
